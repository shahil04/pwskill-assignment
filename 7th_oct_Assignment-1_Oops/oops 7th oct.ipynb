```python
# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. 
# Implement methods to rent a vehicle,
#  return a vehicle, and display available vehicles.
class Vehicle:
    def __init__(self,vehicle_number:str, vehicle_type:str,license:int):
        self.vehicle_number =vehicle_number
        self.vehicle_type =vehicle_type
        self.license =license
        self.is_available =True

    def __str__(self) -> str:
        return f"car number {self.vehicle_number } \ncar type {self.vehicle_type}\nlicense number {self.license} \navailabe {self.is_available}"

class VehicleRentalSystem:
    def __init__(self):
       self.vehicles =[]
    
    def registerVehicle(self,vehicle):
        self.vehicles.append(vehicle)
        

    def availableVehicle(self,vehicle_type):
        for vehicle in self.vehicles:
            if vehicle.vehicle_type ==vehicle_type and vehicle.is_available ==True:
                print('Vehicle is availabe')
                print(vehicle.is_available)
                return vehicle
            

        
    def rentVehicle(self,vehicle_type):
        vehicle = self.availableVehicle(vehicle_type)
        if vehicle.is_available==True:
            print("rent the car successufully")
            vehicle.is_available =False


    def returnVehicle(self,vehicle_number):
        for van in self.vehicles:
            if vehicle_number ==van.vehicle_number and van.is_available==False:
                print("rented")
                van.is_available =True
                print("return successful ")

      
    def __str__(self):
        return str([(van.vehicle_number, van.vehicle_type,van.license ,van.is_available) for van in self.vehicles])


#crate class object and test with value
car =Vehicle('2323','4 wheeler',34)
print(car)
van = Vehicle(1001,'goods load',43)
rent1 =VehicleRentalSystem()
rent1.registerVehicle(car)
rent1.registerVehicle(van)
print(rent1)
# rent1.availableVehicle('goods load')
rent1.rentVehicle('goods load')
print(rent1)
rent1.returnVehicle(1001)
print(rent1)
```

```python
# Problem 4: Library Catalog Create classes representing a library and a book.
#  Implement methods to add books to the library,
# borrow books, and display available books.

class Books:
    '''
    THis class is create book details
    Params: book title
            author name
            isbn number
    '''
    def __init__(self,title:str ,author:str ,isbn: int):
        self.book_title = title 
        self.isbn = isbn
        self.author = author
    
    def __str__(self) -> str:
        return f"title {self.book_title} author {self.author} isbn {self.isbn}"
    

# from Books import get_book_details
from typing import List
class Library:
    '''
    This class do all libraries function 
    '''
    def __init__(self):
        self.book_list =[]

    def addBook(self,book):
        self.book_list.append(book)

    def availableBooks(self,title,author):
        for book in self.book_list:
            if((book.author ==author) and (book.book_title ==title)):
                print(title +" by "+author +" book is availabe ",)
                return book
            
    
    def borrowBook(self,title, author):
        book = self.availableBooks(title, author)
        self.book_list.remove(book)
        print(f"{book} book is assingn to student successfully.\n")
            
    def __str__(self):
        return '\n'.join([str(book) for book in self.book_list])


#DRIVER CODE 
# Test the class and create object
book1 = Books('mindset','shahil',3452323)
book2 = Books('vision','rahis',25645465)
book3 = Books('last moment','shahil',123)
book4 = Books('small goal','muku',1000001)
lib = Library()

lib.addBook(book1)
lib.addBook(book2)
lib.addBook(book3)
lib.addBook(book4)

lib.borrowBook('vision','rahis')
print(lib)
```

```python
# Problem 5: Product Inventory Create classes representing a product and an inventory system.
# Implement methods to add products to the inventory, update product quantity, 
# and display available products.
from typing import List
class Catagory:
    def __init__(self,catagory):
        self.catagory = catagory

    def __str__(self) -> str:
        return f"{self.catagory } catagory"
    
class Product:
    def __init__(self,product_name,price,quantity,catagory):
        self.product_name = product_name
        self.price =price
        self.quantity = quantity
        self.catagory = catagory 
        
    def __str__(self):
        return f"product details { self.product_name} {self.price}rs. {self.quantity}kg {self.catagory}"


class Inventory:
    def __init__(self) -> None:
        self.products :List[Product]=[]

    def addProductInInvetory(self,product):
        self.products.append(product)

    #update the quantity function
    def updateProductQuantity(self,product_name,quantity):
        for product in self.products:
            if product_name in product.product_name:
                print(f"Yes, '{product_name}' is in the inventory.")
                print("in start",product.quantity,"sell",quantity)
                if quantity<0:
                    if product.quantity!=0:
                        if product.quantity< abs(quantity):
                            print(f"product is not enough only {product.quantity} left")
                        else:
                            product.quantity +=quantity
                    else:
                        if product.quantity>quantity:
                            print(f"product is not enough only {product.quantity} left")

                if quantity>0:
                    product.quantity +=quantity
                    
                print("left",product.quantity)
    
    def displayAvailableProduct(self):
        for product in self.products:
            print(product)

food =Catagory('food')
toothpaste =Catagory('toothpaste')
print(food)
print(food.catagory)
rice =Product('baba rice',23 ,12,food.catagory)
rice2 =Product('basmati rice',50 ,120,food.catagory)
rice3 =Product('china rice',100 ,69,food.catagory)
colgate =Product('colgate',17,120,toothpaste.catagory)
print(rice)
invent =Inventory()
invent.addProductInInvetory(rice)
invent.addProductInInvetory(rice2)
invent.addProductInInvetory(rice3)
invent.addProductInInvetory(colgate)
invent.displayAvailableProduct()
invent.updateProductQuantity('baba rice' ,-10)
invent.displayAvailableProduct()
invent.updateProductQuantity('baba rice' ,-12)
invent.updateProductQuantity('baba rice' ,-2)
# # 
invent.updateProductQuantity('baba rice' ,-52)

invent.updateProductQuantity('baba rice' ,50)
```

```python

# Problem 6: Shape Calculation Create a class representing a shape with attributes
#  like length, width, and height.\
#  Implement methods to calculate the area and perimeter of the shape.
class Shape:
    def __init__(self,length, width, height =0) -> None:
        self.length= length
        self.width =width
        self.height =height

class Calculation:
    
    def area(self, shape):
        if shape.length ==shape.width and shape.height==0:
            return f'area of square is {shape.length**2}'
        
        elif shape.length ==shape.width and shape.length ==shape.height:
            return f"area of square cube is {6* shape.length**2}"
        
        elif shape.height !=0:
            return f" area of the rectangular cubiod is {2*shape.height*(shape.length + shape.width)}"
        
        else:
            return f"area of rectangle is {shape.length* shape.width}"


    def perimeter(self, shape):
        if shape.length ==shape.width and shape.height==0:
            return f'Perimeter of square is {4*shape.length}'
        
        elif shape.length ==shape.width and shape.length ==shape.height:
            return f"Perimeter of square cube is {12* shape.length}"
        
        elif shape.height !=0:
            return f" Perimeter of the rectangular cubiod is {4*(shape.length + shape.width + shape.height)}"
        
        else:
            return f"Perimeter of rectangle is {2*(shape.length + shape.width)}"

rect = Shape(2,1,2)
cal = Calculation()
print(cal.area(rect))
print(cal.perimeter(rect))
```

```python
# Problem 7: Student Management Create a class representing a student with
#  attributes like student ID, name, and grades. 
# Implement methods to calculate the average grade and display student details.

class Student:

    def __init__(self,student_id,name,class_name) -> None:
        self.student_id =student_id
        self.name = name
        self.class_name = class_name
        self.marks={}
        self.grade= ''

    def addMarks(self,marks:dict):
        self.marks= marks

    def calculateAverage(self):
        total =0
        for mark in self.marks.values():
            total+=mark
        average =total/len(self.marks)
        percentage =int(average)
        if percentage>=80:
            self.grade='A'
        elif(percentage>=60 and percentage<80):
            self.grade='B'
        else:
            self.grade='C'
        

        return  average,percentage,self.grade

    def displayStudentDetails(self):
        avg,per ,grade =self.calculateAverage()
        return f""" name: {self.name}\n roll no:{self.student_id}\n class: {self.class_name} 
        marks: {self.marks} \n Average: {avg} \nPercentage {per} \nGrade {grade}"""
    
raj =Student(1,'raju','10th')
raj.addMarks({'eng':80,'hindi':40,'maths':40,'sci':40})
raj.calculateAverage()
print(raj.displayStudentDetails())
```

```python
# Problem 8: Email Management Create a class representing an email with attributes like sender, 
# recipient, and subject. Implement methods to send an email and display email details.

class Email:
    def __init__(self, sender, recipient, subject) -> None:
        self.sender =sender
        self.recipient =recipient
        self.subject = subject
        self.email_message =''
        self.is_sent =False

    def message(self,message):
        self.email_message =message

    def send(self):
        self.is_sent =True
        print("Email sent successfully!")


    def display(self):
        status ="sent" if self.is_sent else "Not sent"
        return f"{self.sender ,self.recipient, self.subject ,self.email_message, status}"
    
email=Email("sh@gmail.com","ra@gmail.com","to apply the job")
email.message("i am jus wating for your response .")

print(email.display())
email.send()
email.display()
```

```python
# Problem 9: Social Media Profile Create a class representing a social media profile
#  with attributes like username and posts.
#  Implement methods to add posts, display posts, and search for posts by keyword.

class Profile:
    def __init__(self, username:str, dob:str, phone_no:int,address:str , bio:str, image_url:str):
        self.username =username
        self.dob =dob
        self.phone_no =phone_no
        self.address =address
        self.bio =bio
        self.image_url =image_url

class Post:
    def __init__(self,profile) -> None:
        self.user =profile.username
        self.post_count =0
        self.posts= []
    
    def addPost(self, image_url, message ):
        self.post_id =self.post_count
        self.post_count +=1
        self.posts.append([self.post_id, image_url, message])
        print(f"{self.user} add post successfully ")
    
    def displayPost(self):
        return self.posts
    
    def postSearchByKeyword(self, keyword):
        temp_list =[]
        for message in self.posts:
            if keyword in message[2]:
                temp_list.append(message)
        return temp_list

    
raj =Profile('raj','23/1/12',9874155855,'ranchi','Student at pwskills','https://img.com')
post = Post(raj)
post.addPost("http:posturl.com"," hi everyone")
post.addPost("http:posturl1.com"," hi everyone new message")
post.addPost("http:posturl121.com"," new message day 3")
print(post.displayPost())
print(post.postSearchByKeyword("hi"))
```

```python
# Problem 10: ToDo List Create a class representing a ToDo list with attributes 
# like tasks and due dates. 
# Implement methods to add tasks, mark tasks as completed, and display pending tasks.
from datetime import datetime 
class ToDoList:

    def __init__(self) -> None:
        self.no_of_task:int =0
        self.remaining_task=0
        self.completed_task=0
        self.due_dates=datetime.today()
        self.tasks =[]
        self.date = datetime.today()
        self.mark =False

    def addTask(self, task):
        self.task =task
        self.mark =False
        self.no_of_task +=1
        self.remaining_task +=1
        self.tasks.append([self.no_of_task,self.task,self.mark])

    def markTask(self,taskNumber:int):
        for task in self.tasks:
            if task[0]==taskNumber:
                task[2] =True
                self.remaining_task -=1
                self.completed_task +=1
                
    def displayPendingTask(self):
        print("Pending Task\n")
        for task in self.tasks:
            if task[2]==False:
                print(task)

    def displayTasksDetails(self):
        print(f' {self.date.strftime("%A %d %B %Y")}')
        print(f"Total task {self.no_of_task}\n=====================")
        for task in self.tasks:
            print({task[0]}, task[1])
        print(f"======================\n Completed Task {self.completed_task}")
        for task in self.tasks:
            if task[2]==True:   
                print(f"{task[0]} {task[1]} ===>Done")
        print(f"======================\n Remaining Task ")
        for task in self.tasks:
            if task[2]==False:
                print(f"{task[0]} {task[1]} ==>______")

obj =ToDoList()
obj.addTask('exercise,meditation')
obj.addTask('2hr DSA code ')
# obj.displayPendingTask()
obj.markTask(1)
# obj.displayPendingTask()
obj.displayTasksDetails()


```
